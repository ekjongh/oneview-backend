```python
# Group code: 1: 조, 2: 팀, 3: 센터, 4: 본부
# Date code: 1: 일, 2: 주, 3: 월
def get_voc_group_by_group_date(filter_group, start_date=None, end_date=None):
    voc_cnt = func.sum(models.VocList.wjxbfs1).label("voc_cnt")
    juso = func.concat(models.VocList.sido_nm, models.VocList.eup_myun_dong_nm).label("juso")
    entities = [
        models.VocList.base_ym,
        models.VocList.base_date,
        models.VocList.equip_cd0,
        juso,
        models.VocList.bts_oper_team_nm,
        models.VocList.adm_cd1,
        voc_cnt
    ]
    stmt = select(*entities)
    
    if not end_date:
        end_date = start_date
        
    if start_date:
        stmt = stmt.where(between(models.VocList.base_date, start_date, end_date))
    
    if filter_group:
        stmt = stmt.where(models.VocList.bts_oper_team_nm == filter_group)
    
    stmt = stmt.group_by(models.VocList.base_ym,
                         models.VocList.base_date,
                         models.VocList.equip_cd0,
                         juso,
                         models.VocList.bts_oper_team_nm,
                         models.VocList.adm_cd1).order_by(voc_cnt.desc())
    return stmt
```

```python
rst = get_voc_group_by_group_date(filter_group="구로엔지니어링부", start_date="20220511", end_date="20220521")
print(db.execute(rst).fetchmany(size=10)[0])
```
